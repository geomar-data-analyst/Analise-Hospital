## 18. Importa√ß√£o das bibliotecas de gr√°ficos üß∞

```python
# üß∞ Importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Estilo dos gr√°ficos
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)

# üóÇÔ∏è Carregando o dataset
df_hospital = pd.read_csv('hospital.csv')
```

## 19. Gr√°ficos da an√°lise do sexo dos pacientes üìä

```python
sns.countplot(x='sexo', data=df_hospital, palette='pastel')
plt.title('Distribui√ß√£o de Pacientes por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Quantidade')
plt.show()
```

## 20. Tipos de interna√ß√£o üè•

```python
sns.countplot(x='tipo_internacao', data=df_hospital, palette='Set2')
plt.title('Tipos de Interna√ß√£o')
plt.xlabel('Tipo')
plt.ylabel('Quantidade')
plt.show()
```

## 21. Tempo m√©dio de interna√ß√£o üìÖ

```python
sns.boxplot(x='tipo_internacao', y='dias_internacao', data=df_hospital, palette='Set3')
plt.title('Dias de Interna√ß√£o por Tipo')
plt.xlabel('Tipo de Interna√ß√£o')
plt.ylabel('Dias Internado')
plt.show()
```

## 22. Diagn√≥sticos mais frequentes ü¶†

```python
top_diag = df_hospital['diagnostico_principal'].value_counts().head(10)
sns.barplot(x=top_diag.values, y=top_diag.index, palette='coolwarm')
plt.title('Top 10 Diagn√≥sticos')
plt.xlabel('Quantidade')
plt.ylabel('Diagn√≥stico')
plt.show()
```

## 23. Tratamentos mais realizados üíâ

```python
top_trat = df_hospital['tratamento_principal'].value_counts().head(10)
sns.barplot(x=top_trat.values, y=top_trat.index, palette='viridis')
plt.title('Top 3 Tratamentos')
plt.xlabel('Quantidade')
plt.ylabel('Tratamento')
plt.show()
```

## 24. Especialidades m√©dicas mais acionadas ü©∫

```python
esp = df_hospital['especialidade_medica'].value_counts()
sns.barplot(x=esp.values, y=esp.index, palette='mako')
plt.title('Especialidades M√©dicas Mais Demandadas')
plt.xlabel('Quantidade')
plt.ylabel('Especialidade')
plt.show()
```

## 25. Condi√ß√£o dos pacientes ‚ö∞Ô∏è

```python
sns.countplot(x='condicao_saida', data=df_hospital, palette='rocket')
plt.title('Condi√ß√£o de Sa√≠da dos Pacientes')
plt.xlabel('Condi√ß√£o')
plt.ylabel('Quantidade')
plt.show()
```

## 26. Custo m√©dio por plano de sa√∫de üí∞

```python
custos = df_hospital.groupby('plano_saude')['custo_total'].mean().sort_values()
sns.barplot(x=custos.values, y=custos.index, palette='flare')
plt.title('Custo M√©dio por Plano de Sa√∫de')
plt.xlabel('Custo M√©dio (R$)')
plt.ylabel('Plano de Sa√∫de')
plt.show()
```

## 27. Satisfa√ß√£o m√©dica por especialidade üòä

## 28. Taxa de reinterna√ß√£o em 30 dias üîÅ

## 29. Correla√ß√£o entre dias internado e custo üìà
